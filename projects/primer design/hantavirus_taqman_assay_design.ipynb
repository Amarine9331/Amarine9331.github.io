## Hantavirus Biology in One Page

Hantaviruses are negative-sense RNA viruses in the order *Bunyavirales*.
Human disease is associated with rodent-borne orthohantaviruses. The
International Committee on Taxonomy of Viruses places disease-associated
hantaviruses in the family *Hantaviridae*, subfamily *Mammantavirinae*,
genus *Orthohantavirus*.

Orthohantaviruses have three genome segments:

- **S segment:** encodes nucleoprotein N, the abundant structural
  protein that encapsidates viral RNA.
- **M segment:** encodes the glycoprotein precursor, cleaved into Gn and
  Gc, which mediate entry.
- **L segment:** encodes the viral RNA-dependent RNA polymerase.

From a diagnostic target-selection perspective:

- The **S segment** is often attractive because nucleoprotein-coding
  regions can be relatively conserved within a virus species and
  abundant transcripts may aid detection.
- The **M segment** can provide lineage information but may show greater
  diversity in glycoprotein regions under host or immune selection.
- The **L segment** contains conserved polymerase motifs that may
  support broad detection but can be less discriminating among related
  viruses.
- Segmented genomes create reassortment risk, so a robust assay strategy
  may use more than one target if a clinical or public-health decision
  depends on lineage certainty.

## Epidemiology and Host Ecology

Hantavirus risk is shaped by host ecology. Human infection classically
follows inhalation of aerosolized excreta or secretions from infected
rodents or direct rodent contact. Geographic patterns reflect the
distribution of reservoir hosts.

Broad clinical-epidemiological groupings are:

| Group | Typical geography | Examples | Main syndrome association |
|-----------------|---------------------:|-----------------|-----------------|
| Old World orthohantaviruses | Europe and Asia | Hantaan, Dobrava, Puumala, Seoul | Hemorrhagic fever with renal syndrome |
| New World orthohantaviruses | Americas | Andes, Sin Nombre | Hantavirus pulmonary/cardiopulmonary syndrome |

For the MV Hondius case study, the key epidemiological feature is not
that hantaviruses suddenly became broadly transmissible. Rather, Andes
virus has a documented history of limited person-to-person transmission,
and a cruise ship is a closed travel environment where contact networks,
shared cabins, caregiving, medical evacuation, and international
disembarkation can complicate response.

## Phylogeny: What the Tree Should Answer

Phylogenetics should serve an assay-development question, not just
produce a tree. For a hypothetical Andes virus TaqMan assay, the useful
questions are:

1.  **Target scope:** Should the assay detect all known Andes virus
    diversity, a recent outbreak lineage, or a broader American
    hantavirus group?
2.  **Near-neighbor risk:** Which related orthohantaviruses are close
    enough to create cross-reactivity concerns?
3.  **Sequence coverage:** Are available genomes representative of
    geography, time, host, and clinical cases?
4.  **Segment consistency:** Do S, M, and L trees tell the same story,
    or is reassortment or incomplete sampling plausible?
5.  **Outbreak stability:** Do the outbreak sequences introduce
    mismatches in any candidate binding region?


## Diagnostic Design Goal

The hypothetical product requirement is:

> Develop a TaqMan-style one-step RT-qPCR screening assay intended to
> detect Andes virus RNA in appropriately handled clinical or
> public-health specimens, with design evidence supporting broad
> inclusivity across known Andes virus diversity and exclusivity against
> near-neighbor hantaviruses, common respiratory pathogens, host
> sequences, and environmental/background organisms.

This is a design brief, not a validated intended-use statement. A real
clinical assay would require regulatory review, biosafety assessment,
specimen-type validation, analytical and clinical performance studies,
controls, quality systems, and laboratory authorization.

## Why TaqMan RT-qPCR?

A hydrolysis-probe RT-qPCR format is commonly used in molecular
diagnostics because it can combine:

- sequence-specific amplification and sequence-specific probe detection,
- real-time amplification monitoring,
- compatibility with multiplex internal controls,
- straightforward analytical validation metrics such as limit of
  detection, precision, interference, inclusivity, and cross-reactivity.

For an RNA virus, reverse transcription must precede amplification. From
a diagnostic-development perspective, the design risk is not just
whether a primer pair can amplify a reference sequence; it is whether
the assay remains robust across target diversity and does not produce
misleading positives from non-target templates.

## Target Selection Strategy

Candidate targets should be evaluated across the three segments:

| Criterion | S segment | M segment | L segment |
|------------------|------------------|------------------|------------------|
| Within-Andes conservation | Often favorable | Variable by region | Conserved motifs possible |
| Discrimination from near neighbors | Moderate to high | Often high | May be lower in conserved motifs |
| Phylogenetic information | Good for lineage context | Good for lineage/context | Good for deeper relationships |
| Diagnostic risk | Misses if local variation affects binding | Misses if highly variable | Cross-reactivity if too conserved |

Pragmatic strategy:

1.  Start with conserved windows in S and L.
2.  Reject windows with known outbreak-sequence mismatches near primer
    3-prime ends or probe-critical positions.
3.  Prefer windows that are conserved across target diversity but
    separated from near-neighbor hantaviruses by multiple informative
    differences.
4.  Consider a two-target assay if single-target dropout risk is
    unacceptable.

## In-Silico Primer/Probe Design Constraints

The following are design-review constraints, not a complete recipe:

| Feature | Review expectation |
|------------------------------------|------------------------------------|
| Amplicon | Short enough for degraded RNA and efficient qPCR, while retaining specificity |
| Primers | Similar melting behavior, limited self-complementarity, limited primer-dimer risk |
| Probe | Binds within a conserved, target-informative region and avoids highly variable positions |
| 3-prime ends | Avoid target polymorphisms and avoid strong non-target complementarity |
| GC content | Avoid extremes that impair binding, amplification, or synthesis |
| Degeneracy | Use sparingly; degeneracy can improve inclusivity but reduce sensitivity and specificity |
| Multiplexing | Check interaction among all primers, probes, and internal controls |
| Sequence drift | Establish a surveillance process for new public genomes |

No candidate oligonucleotide sequences are shown here. For a portfolio
piece, the strongest demonstration is a transparent decision framework,
ranked candidate windows, and a validation plan rather than a deployable
assay.

## Inclusivity Testing

**Inclusivity** asks whether target strains or target nucleic acids are
expected to test positive. In silico, the panel should cover the
intended target breadth.

For an Andes virus assay, the inclusivity panel should include:

- geographically diverse Andes virus sequences,
- human and reservoir-associated sequences where available,
- historical and recent sequences,
- all available segments relevant to the chosen target,
- outbreak-associated sequences treated as a challenge set,
- partial sequences only when the candidate target region is covered and
  quality is sufficient.

In-silico inclusivity outputs should include:

- number of target sequences evaluated,
- number with full amplicon coverage,
- number with perfect primer and probe matches,
- mismatch count by oligo and position,
- special flag for mismatches near primer 3-prime ends,
- estimated dropout risk by lineage or geography,
- a list of sequences excluded because of incomplete coverage or poor
  quality.

``` python
# Mock schema for an inclusivity review table.
# Values below are placeholders for notebook demonstration only.

import pandas as pd

inclusivity_schema = pd.DataFrame(
    [
        {
            "sequence_id": "ANDV_reference_or_lineage_id",
            "segment": "S",
            "region": "candidate_window_A",
            "forward_mismatches": "count_and_positions",
            "probe_mismatches": "count_and_positions",
            "reverse_mismatches": "count_and_positions",
            "risk_call": "pass / review / fail",
            "rationale": "e.g., no critical 3-prime mismatch",
        }
    ]
)

inclusivity_schema
```

### Inclusivity Decision Rules

An assay-development team should define decision rules before ranking
candidates. Example rules:

- **Pass:** all target sequences with adequate coverage have no critical
  primer/probe mismatches.
- **Review:** one or more target sequences have mismatches outside
  critical positions, or coverage gaps make a lineage hard to assess.
- **Fail:** a known target lineage or outbreak sequence has a mismatch
  pattern expected to reduce amplification or probe detection.

These rules should be documented and version controlled. The decision
should not be adjusted after seeing inconvenient results unless the
change is justified and recorded.

## Exclusivity Testing

**Exclusivity** asks whether non-target organisms or nucleic acids are
expected to test negative. For a molecular assay, the panel should
include both phylogenetic near neighbors and realistic sample-background
organisms.

For Andes virus, exclusivity should include:

- non-Andes New World hantaviruses,
- Old World hantaviruses,
- common respiratory viruses that may be present in respiratory
  specimens,
- human genome and transcriptome sequences,
- rodent host sequences if environmental or reservoir samples are in
  scope,
- bacterial and fungal organisms common in respiratory or environmental
  backgrounds, if those matrices are claimed.

In-silico exclusivity should not rely only on short local matches. The
critical question is whether a primer pair can plausibly amplify a
product of the expected size and whether the probe can bind that
product. NCBI Primer-BLAST is useful because it combines primer design
with specificity checking; NCBI advises selecting the relevant organism
and smallest appropriate database for precise checks, or broader
databases when broad coverage is needed.

Suggested ranking components:

- **Target conservation:** fraction of target sequences without critical
  mismatches.
- **Near-neighbor separation:** absence of plausible non-target
  amplicons.
- **Oligo physical quality:** absence of strong self-complementarity,
  primer dimers, extreme GC, or problematic runs.
- **Outbreak robustness:** no critical mismatch in current outbreak
  sequences.
- **Manufacturability:** avoids unusual modifications unless justified.
- **Change-control burden:** target region is stable enough for
  surveillance.

## Analytical Validation Plan

After in-silico review, a real molecular diagnostic assay would need
laboratory validation. This notebook does not provide protocols, but the
validation plan would normally address:

| Performance characteristic | Question |
|------------------------------------|------------------------------------|
| Limit of detection | What is the lowest concentration detected with acceptable probability in each claimed specimen type? |
| Inclusivity | Does the assay detect representative target diversity? |
| Exclusivity / cross-reactivity | Does the assay avoid detection of near neighbors and likely co-occurring organisms? |
| Precision | Are results reproducible across operators, instruments, reagent lots, days, and sites? |
| Interference | Do substances or specimen matrix effects inhibit detection? |
| Stability | Are specimens, controls, and reagents stable under claimed conditions? |
| Controls | Do extraction, amplification, inhibition, positive, and negative controls behave as expected? |
| Clinical evaluation | Does assay performance support the intended use in the target population? |

CLSI MM03 describes molecular diagnostic methods for infectious
diseases, including sequence selection, performance evaluation,
inhibitors, false-positive control, interpretation, quality assurance,
and regulatory considerations. CLSI MM17 addresses validation and
verification concepts for qualitative multiplex nucleic acid assays.
Those standards are useful anchors for turning an in-silico design into
a regulated diagnostic-development plan.

## Regulatory and Quality Considerations

An industry or clinical laboratory team should separate four claims:

1.  **In-silico design claim:** the target window is predicted to be
    inclusive and exclusive based on the current sequence database.
2.  **Analytical performance claim:** the assay detects target material
    under controlled laboratory conditions.
3.  **Clinical performance claim:** the assay identifies infected
    patients in the intended population and specimen type.
4.  **Public health utility claim:** the assay supports outbreak
    response, triage, surveillance, or contact investigation.

Only the first claim can be supported by this notebook. The other claims
require laboratory and clinical evidence.

Version control should capture:

- source databases and access dates,
- accession lists and exclusion criteria,
- alignment and tree software versions,
- primer/probe design software and parameters,
- candidate scoring rules,
- review sign-off,
- periodic re-analysis triggers when new sequences appear.

## Failure Modes and Mitigations

| Failure mode | How it could happen | Mitigation |
|------------------------|------------------------|------------------------|
| False negative from sequence drift | Mutation in primer/probe binding site | Multi-sequence inclusivity screen and periodic surveillance |
| False negative from incomplete sampling | Missing lineage absent from databases | Conservative target choice and documented uncertainty |
| False positive from near neighbor | Related hantavirus amplified | Near-neighbor exclusivity panel and in-silico PCR |
| False positive from contamination | Amplicon or control contamination | Workflow controls and physical separation in real labs |
| Inhibition | Specimen matrix suppresses amplification | Internal control and specimen validation |
| Overclaiming | In-silico pass treated as clinical validation | Clear claim boundaries and regulatory review |

## Discussion

The MV Hondius cluster is a strong case study because it shows how
modern molecular diagnostics sits between public health and molecular
biology. Phylogeny helps define the target and near-neighbor space.
Epidemiology defines who might be tested, what specimens might be
relevant, and what turnaround time would matter. Assay design translates
sequence conservation into a practical screen, while inclusivity and
exclusivity testing force the developer to state what the assay is, and
is not, designed to detect.

The most important scientific discipline in this exercise is restraint.
A sequence-perfect candidate is not a clinical test. A beautiful
phylogenetic tree does not prove transmission. A negative in-silico
cross-reactivity search does not replace wet-lab validation. For
molecular diagnostics, evidence must be layered: genomic rationale
first, analytical performance second, clinical performance third, and
public-health utility last.
